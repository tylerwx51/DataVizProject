```python
import pandas as pd
```

```python
marvel = pd.read_csv('marvel_comic.csv')
dc = pd.read_csv('dc_comic.csv')
```

```python
marvel.columns = ['page_id','name','urlslug','id','align','eyes','hair','sex','gsm','alive','appearances','first appearance','year']
dc.columns = ['page_id','name','urlslug','id','align','eyes','hair','sex','gsm','alive','appearances','first appearance','year']
```

```python
marvel.drop('first appearance', axis = 1, inplace = True)
dc.drop('first appearance', axis = 1, inplace = True)
```

### align

```python
dc['align'] = [str(x).replace('Reformed Criminals','nan') for x in dc['align']]
```

```python
set(dc['align'])
```

### name

```python
set(marvel['name'])
```

```python
marvel['name'] = marvel['name'].map(lambda x: x.replace(' (Earth-616)',''))
```

```python
set(dc['name'])
```

```python
dc['name'] = dc['name'].map(lambda x: x.replace(' (New Earth)',''))
```

### eye

```python
marvel['eyes'] = [str(x).replace('Eyeballs','Eyes') for x in marvel['eyes']]
dc['eyes'] = [str(x).replace('Auburn Hair','nan') for x in dc['eyes']]
```

```python
marvel['eyes'].value_counts()
```

### hair

```python
marvel['hair'] = [str(x).replace('No Hair','Bald') for x in marvel['hair']]
marvel['hair'] = [str(x).replace('Reddish Blond Hair','Reddish Hair') for x in marvel['hair']]
marvel['hair'] = [str(x).replace('Yellow Hair','Blond Hair') for x in marvel['hair']]
marvel['hair'] = [str(x).replace('Light Brown Hair','Brown Hair') for x in marvel['hair']]
marvel['hair'] = [str(x).replace('Orange-brown Hair','Orange Hair') for x in marvel['hair']]


dc['hair'] = [str(x).replace('Reddish Brown Hair','Reddish Hair') for x in dc['hair']]
dc['hair'] = [str(x).replace('Violet Hair','Purple Hair') for x in dc['hair']]
dc['hair'] = [str(x).replace('Platinum Blond Hair','Blond Hair') for x in dc['hair']]
```

```python
marvel['hair'].value_counts()
```

```python
dc['hair'].value_counts()
```

```python
set(marvel['hair']) - set(dc['hair'])
```

```python
set(dc['hair']) - set(marvel['hair'])
```
